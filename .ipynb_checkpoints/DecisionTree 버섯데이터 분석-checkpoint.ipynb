```python
import pandas as pd
pd.set_option('display.max_columns', 1000)
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
```

```python
# 1. 데이터 로딩
data = pd.read_csv("data/mushroom.csv")
data
```

```python
data.info()
```

```python
X = data.iloc[:, 1:]
y = data.loc[:, 'poisonous']
```

```python
data.loc[:, 'poisonous']
```

```python
data.iloc[:, 1:]
```

```python
print(X.shape)
print(y.shape)
```

```python
data.describe()
```

```python
y.value_counts()
```

## 원 핫 인코딩

```python
X_one_hot = pd.get_dummies(X)  # 글자타입만 원핫인코딩해주는 get_dummies함수
# 데이터가 숫자형이면 원핫인코딩이 제대로 안될 수 있으니 안될 경우 데이터타입을 재확인할 것
X_one_hot
```

```python
X_one_hot.shape
```

## 라벨인코딩

```python
# mapping
dict_poison={"p":1, "e":0}
y.map(dict_poison)
```

## 모델링

```python
X_train, X_test, y_train, y_test = train_test_split(X_one_hot, y, test_size = 0.3, random_state = 65)
y_test
```

```python
tree_model = DecisionTreeClassifier()
tree_model.fit(X_train, y_train)
```

```python
# 모델 스스로 정확도를 판단할 수 있는 함수 score == accuracy_score
tree_model.score(X_test, y_test)
```

#### 학습시킨 decision tree 모델 .dot 그림으로 표현

```python
from sklearn.tree import export_graphviz   #만드는 파일이 .dot파일 형식
```

```python
export_graphviz(tree_model, # 시각화시키고 싶은 모델
               out_file='tree.dot', # 저장할 파일명
               class_names=['식용', '독'], # 클래스이름
               feature_names=X_one_hot.columns, # 학습에 사용된 특성이름
               impurity=True, #불순도시각화여부
               filled=True) # 각 트리 색구분 여부
```

```python
import graphviz
with open("tree.dot", encoding="utf-8") as f:
    dot_graph = f.read()
    
display(graphviz.Source(dot_graph))
```

### 특성중요도

```python
tree_models = DecisionTreeClassifier(max_depth=2)
# max_depth  깊이
# max_leaf_nodes  노드 수
# min_samples_leaf  노드별 최소 요소수
tree_models.fit(X_train, y_train)
```

```python
export_graphviz(tree_models, # 시각화시키고 싶은 모델
               out_file='tree.dot', # 저장할 파일명
               class_names=['식용', '독'], # 클래스이름
               feature_names=X_one_hot.columns, # 학습에 사용된 특성이름
               impurity=True, # 지니불순도 시각화여부
               filled=True) # 각 트리 색구분 여부
```

```python
import graphviz
with open("tree.dot", encoding="utf-8") as f:
    dot_graph = f.read()
    
display(graphviz.Source(dot_graph))
```
